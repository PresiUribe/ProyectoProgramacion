```python
# Se define el intervalo de tiempo que vamos a trabajar, este caso el mes de diciembre.
# Tambien pasamos la hora, que venia en el formato UTC, a la zona horario de America latina / Bogota.
dic = "2023-12-01"
dic1 = "2023-12-31"
df['date'] = pd.to_datetime(df['date'], utc=True)
df['date'] = df['date'].dt.tz_convert('America/Bogota')
datosfil = df[(df['date'] >= dic) & (df['date'] <= dic1)]
```
