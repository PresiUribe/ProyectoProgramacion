```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/gdrive')
```

```python
df = pd.read_excel('/gdrive/MyDrive/proyecto/Report_CD 61283 TR1 _2024-04-15 (2).xlsx')
```

```python
df.shape
```

```python
# Se define el intervalo de tiempo que vamos a trabajar, este caso el mes de diciembre.
# Tambien pasamos la hora, que venia en el formato UTC, a la zona horario de America latina / Bogota.
ini = "2023-12-01"
fin = "2023-12-31"
df['date'] = pd.to_datetime(df['date'], utc=True)
df['date'] = df['date'].dt.tz_convert('America/Bogota')
datosfil = df[(df['date'] >= ini) & (df['date'] <= fin)]
```

```python
# Se filtran los datos para el circuito 1 o Region 1.
# Se sacan los datos de las columnas IA, IB, IC, ademas de las fechas para tales datos.
dtfil1 = datosfil.loc[datosfil.name == "Circuito I"]
iar1 = dtfil1.IA
ibr1 = dtfil1.IB
icr1 = dtfil1.IC
t1 = dtfil1.date
```

```python
#######################################
plt.figure(figsize=(15, 5))
plt.plot(t1, iar1, marker='o')
plt.title('Corriente IA de la Region 1')
plt.xlabel('Tiempo (h)')
plt.ylabel('Corriente IA (A)')
plt.grid(True)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
#######################################
plt.figure(figsize=(15, 5))
plt.plot(t1, ibr1, marker='o')
plt.title('Corriente IB de la Region 1')
plt.xlabel('Tiempo (h)')
plt.ylabel('Corriente IB (A)')
plt.grid(True)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
#######################################
plt.figure(figsize=(15, 5))
plt.plot(t1, icr1, marker='o')
plt.title('Corriente IC de la Region 1')
plt.xlabel('Tiempo (h)')
plt.ylabel('Corriente IC (A)')
plt.grid(True)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Se extrae la columna kWh del dataframe por regiones para establecer los consumos
consumo = dtfil1.kwh
plt.figure(figsize=(8, 6))
plt.boxplot(consumo.dropna())
plt.title('Boxplot de la los kWh de la Region 1')
plt.ylabel('Valores')
plt.grid(True)

plt.show()
```

```python
# Se busca el promedio de horas  por regin para graficar la cantidad de consumo en cada franja horaria.
promhorar1 = dtfil1.groupby(dtfil1['date'].dt.hour)['kwh'].mean()
promhorar2i = dtfil2.groupby(dtfil2['date'].dt.hour)['kwh'].mean()
promhorar3ii = dtfil3.groupby(dtfil3['date'].dt.hour)['kwh'].mean()

plt.figure(figsize=(14, 7))

plt.subplot(1, 3, 1)
promhorar1.plot(kind='bar', color='royalblue')
plt.title('Region I')
plt.xlabel('Hora del Da (Hora UTC-5)')
plt.ylabel('Consumo Medio (kWh)')

plt.subplot(1, 3, 2)
promhorar2i.plot(kind='bar', color='gold')
plt.title('Region II')
plt.xlabel('Hora del Da (Hora UTC-5)')
plt.ylabel('Consumo Medio (kWh)')

plt.subplot(1, 3, 3)
promhorar3ii.plot(kind='bar', color='lime')
plt.title('Region III')
plt.xlabel('Hora del Da (Hora UTC-5)')
plt.ylabel('Consumo Medio (kWh)')

plt.tight_layout()
plt.show()
```
